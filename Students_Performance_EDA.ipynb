## ðŸ“Š GPA Distribution

Understanding the distribution of GPA helps identify overall student performance trends.


```python
sns.histplot(data['GPA'], kde=True)
plt.title("GPA Distribution")
plt.show()
```

## ðŸ“Š Weekly Study Time Distribution

Understanding the distribution of Weekly Study Time helps identify overall student performance trends.


```python
sns.histplot(data['StudyTimeWeekly'], kde=True)
plt.title("Weekly Study Time")
plt.show()
```

**Insight:**  
Most students have a GPA concentrated around the average range, with fewer high-performing and low-performing students.


## ðŸ“Š Study Time vs GPA


```python
sns.scatterplot(x='StudyTimeWeekly', y='GPA', data=data)
plt.title("Study Time vs GPA")
plt.show()
```

## ðŸ“Š Absences vs GPA


```python
sns.scatterplot(x='Absences', y='GPA', data=data)
plt.title("Absences vs GPA")
plt.show()
```

## ðŸ“Š Extracurricular Activities and GPA


```python
activities = ['Sports', 'Music', 'Volunteering', 'Extracurricular']

for col in activities:
    sns.boxplot(x=col, y='GPA', data=data)
    plt.title(f"GPA by {col}")
    plt.show()
```

## ðŸ”— Correlation Analysis


```python
num_cols = data.select_dtypes(include='number')

plt.figure(figsize=(10,6))
sns.heatmap(num_cols.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

**Insight:**  
Study time has a positive correlation with GPA, while absences show a negative correlation.


## ðŸ§  Key Findings

- Study time is one of the strongest indicators of academic success.
- Absenteeism negatively affects GPA.
- Extracurricular activities have a modest positive impact.
